# Week 0 Report: GBM Simulations for Option Pricing

This notebook simulates stock price paths using Geometric Brownian Motion (GBM)
to understand randomness, risk, and future price uncertainty.

```python
# import libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# set model parameters
S0 = 100        # initial stock price
mu = 0.08      # expected annual return
sigma = 0.2    # volatility
T = 1           # time horizon (1 year)
dt = 1/252      # daily time step
N = int(T/dt)   # number of steps
paths = 10      # number of simulated paths
```

```python
# initialize price matrix
S = np.zeros((paths, N))
S[:, 0] = S0
```

```python
# GBM simulation loop
for i in range(1, N):
    Z = np.random.normal(0, 1, paths)
    S[:, i] = S[:, i-1] * np.exp(
        (mu - 0.5 * sigma**2) * dt + sigma * np.sqrt(dt) * Z
    )
```

```python
# plot simulated price paths
plt.figure(figsize=(12,6))
for i in range(paths):
    plt.plot(S[i])
plt.title("GBM Simulated Price Paths")
plt.xlabel("Time Steps")
plt.ylabel("Price")
plt.show()
```

```python
# histogram of final prices
plt.figure(figsize=(8,5))
plt.hist(S[:, -1], bins=20)
plt.title("Distribution of Final Prices")
plt.xlabel("Final Price")
plt.ylabel("Frequency")
plt.show()
```

First before the relation between randomness and risk, what is randomness? Randomness basically means that the outcomes cannot be predicted with certainty. In many aspects like finance, nature, technology, or daily life, randomness shows that the world does not move in straight, predictable lines. Instead, it shifts in ways we cannot fully predict. Understanding randomness does not eliminate uncertainty, but it helps in understanding the nature of uncertaintyand that is what risk truly is.
Risk exists because the future is unpredictable. Randomness tells us that even when we know the rules of a system, we cannot always know the results. For example, in stock markets, we may understand long-term trends, but day-to-day movements are influenced by countless unseen factors. A random shocknews, emotion, technology failure, or global eventscan instantly change the expected outcome. By studying randomness through models like probability distributions or simulations, we begin to see that outcomes are not just good or bad, but spread over a range. This spread is what we call risk.
Randomness also teaches us that extreme outcomes, though rare, are always possible. Most days may be stable, but occasional large fluctuations carry significant consequences. This is why we consider not just average behavior but also volatility and tail events. In simple terms: randomness warns us that unusual events are not impossiblethey are simply infrequent, and therefore dangerous to ignore.
Most importantly, randomness highlights the limits of human prediction. It reminds us that confidence is not certainty. When we plan for the futureespecially in stock marketswe must prepare for variability, not just the expected path. Good risk management does not try to eliminate randomness; it tries to adapt to it by building buffers, diversifying choices, and accepting uncertainty as part of reality.
In conclusion, randomness teaches us that risk is not something mysterious; it is simply the natural consequence of an unpredictable world. By understanding how randomness shapes outcomes, we understand risk more honestly, prepare more wisely, and make decisions that acknowledge uncertainty instead of ignoring it.
