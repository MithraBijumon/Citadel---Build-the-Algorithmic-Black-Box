```python
import numpy as np
import matplotlib.pyplot as plt

# parameters
S0 = 100 # starting price
mu = 0.08 # expected annual return (drift)
sigma = 0.2 # std
T = 1           # time period = 1 year
dt = 1/252 # daily time step
N = int(T/dt) # number of time steps in the year
paths = 10 # generate 10 simulated price paths

# storage
S = np.zeros((paths, N))
S[:, 0] = S0

for i in range(1, N):
    Z = np.random.normal(0, 1, paths)
    S[:, i] = S[:, i-1] * np.exp((mu - 0.5*sigma**2)*dt + sigma*np.sqrt(dt)*Z)

# Plot paths
plt.figure(figsize=(12,6))
for i in range(paths):
    plt.plot(S[i])
plt.title("GBM Simulated Price Paths")
plt.xlabel("Time Steps")
plt.ylabel("Price")
plt.show()

# Histogram of final prices
plt.figure(figsize=(8,5))
plt.hist(S[:, -1], bins=20)
plt.title("Distribution of Final Prices")
plt.xlabel("Final Price")
plt.ylabel("Frequency")
plt.show()
```
